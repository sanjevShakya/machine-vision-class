```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
point1 = np.array([2, 4, 2])
point2 = np.array([6,3,3])
point3 = np.array([1,2,0.5])
point4 = np.array([16, 8, 4])
line = np.array([8, -4, 0]).T
print(line)
def compute_dot_product(line, point):
    return line.T.dot(point) 
```

```python
p1 = compute_dot_product(line, point1)
p2 = compute_dot_product(line, point2)
p3 = compute_dot_product(line, point3)
p4 = compute_dot_product(line, point4)
dot_points = [p1, p2, p3, p4]
points = [point1, point2, point3, point4]
```

```python
print(points)
```

```python
for i, result in enumerate(dot_points):
    if(result <= 0.0):
        print("Point ({},{}) is in the given line".format(points[i][0]/points[i][2], points[i][1]/points[i][2]))
```

```python
homogenous_points = np.array(points)
# homogenous_points = np.concatenate(homogenous_points, axis=2)
X = homogenous_points[:,0] / homogenous_points[:,2]
Y = homogenous_points[:,1] / homogenous_points[:,2]
# line
def line(x):
    return 2 * x
x_line_points = np.linspace(0, 5, 50)
y_line_points = line(x_line_points)
plt.scatter(X, Y)
plt.plot(x_line_points, y_line_points, 'b-')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.title('Points intersecting a line');
plt.show()
```
